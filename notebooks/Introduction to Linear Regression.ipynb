What is linear regression?
--------------------------

Lets start from the basics.

Remember the equation of a straight line from high school 

$$Y = mX + b$$

where m is the slope and b is the y intercept.  

### Insights from the high school physics lab

Remember also a typical physics lab experiment from middle school. We had some input X (say force) which gave some output Y (say acceleration).  

You made a number of pairs of observations x, y and then you had to fit a straight line through the set of observations using a visual 
"best fit".

And then you read off 'm' and 'b' from the graph, hoping it was close to the expected answer.

---

#### Regression is this idea - estimating the parameters of a formula, a function or an equation  - taken to very sophisticated levels.

To explain, we have a dataset (the observations) and a model (our guess for a formula that fits the data) and we have to figure out the coefficients of the formula (the m and b in our best fit line) so that the "plot" of the equation fits the data the "best". 

Again, we want to use our data to find coefficients for a formula so that the formula will fit the data the "best".  It could also be considered reverse engineering a formula from the data.

A notion of total error
-----------------------

Back to the physics lab ...

Once you had your visual best fit line and had read off the m and b you probably said something to the effect 

"The data follows a linear equation of the form Y = mX + b where m (slope)=(somenumber) and b (y intercept)=(someothernumber)"

You may recall that the equation is not an exact representation because most probably your data points are not all in a perfect straight line. So there is some error varying from one data point to the next data point. Your visual approach subjectively tried to minimize some intuitive "total error" over all the data.  

What you did was intuitive "Linear Regression". You estimated m and b by the "looks right to me" algorithm.
We will start with thsi intuitive notion and rapidly bring some heavy machinery to bear that will allow us to solve pretty sophisticated problems.

At this point your lab exercise may ask you to approximate what Y will be when X is some number outside the range of your measurements.
Then you use the equation above where m and b are now actual numbers say 2.1 and 0.3 resp i.e the equation is Y = 2.1X + 0.3

This equation is your "model"

And you plug in an X to get a Y.  

This is where you are using your model to predict a value or "extrapolate" i.e you are saying that I didn't use this value of X in my experiment and I don't have it in my data but ....
Based on my model Y = 2.1X + 0.3 if I had used this value in my experiment then I believe I would have got an output Y of approx what the straight line suggests.  
You are also able to say my error is expected to be (some number) so I believe the actual value will lie between Y-error and Y+ error.  When used like this we call the X variable the "predictor" as values of Y are predicted based one values of X.

Formally Linear Regression techniques are just this kind of estimation except that we extend this beyond the informal lab "looks right" approach in multiple ways.

a) we represent the error as a formal mathematical function summed up over all the data points to getthe total error.  There are a number of ways to create a total error function and the result of our estimation depends on what approach we use to set this up.  There is a very commonly used form of the total error which is called "quadratic error".  This form appears in many different fields and is also connected to physics by being a representation of energy of a body and in electrical engineering by being a representation fo the energy in a signal.
So for now we will use an error function that has a quadratic form in it. {{ detailed explanation, links}}

b) Once we have set up a total error function then we use mathematical techniques to minimize this error. The techniques may yield exact solutions in some cases or in some other techniques we use an iterative approach that makes an initial estimate then plugs it in to get better and better estimates. By "better" we mean that the "total error function is getting smaller and smaller"
At some point we say the error is small enough, or we see that it's not getting any better and we decide to stop and read off the m and b.  
So here in a) and b) above we have replaced the "looks right" method by formal math but we have kept it as a single dependent variable Y depending on a single independent variable or "predictor" variable X.

c) Another way the "physics lab" model is extended is in the number of independent variables i.e we have more than one variable on the "X" isde of the equation so our model could be

$$ Y = AX_1 + BX_2 + C $$

or 

$$ Y = AX_1 + BX_2 + CX_3 + D $$

or in matrix form 

$$ Y = \boldsymbol{A}X + B $$

[[ needs some explanation and sidebar about matrix ]]

In all of these cases the basic goal is the same.  To set up a model in all these cases the model is "linear" in that it uses a set of linear equations to approximate the data and we want to estimate the coefficients a,b,c,d .... or the coefficient matrix A and the constant matrix B ... which will give us our estimated "model".

So what is Regression really?


So regression in general is any approach we might use to estimate the coefficients of a model using the data to estimate the coefficients.
Most common approaches use calculus to minimize the total error function - multivariable calculus. We use an iterative approach called Gradient Descent.

We also use matrix inversion techniques using tools like MATLAB to solve the set of linear equations (assuming certain stability conditions i.e. conditions that ensure we avoid wildly oscillating or meaningless solutions).

### Linear Regression

This is what we've been looking at so far. We used a linear model i.e. 

$$ Y = Ax + B $$.  
We do this when we explore the data visually and it roughly "looks like" a straight line.  Even if not exactly so, it may still be valuable to see what a "linear approximation" to the data might look like. In many data science explorations we start off with a linear model as our first shot at making sense of the data and it is a vital tool in the toolbox of a data scientist.  

It might be safe to say that without linear regression you are missing a huge essential skill for being an effective data scientist. It might not be an exaggeration to say that Linear Regression and the concepts in it are foundational to a large part of data science especially because they are generalized in many directions to create many more powerful tools and techniques well beyond models that look like a straight line.

### Quadratic Regression

Perhaps after some initial exploration we see that the data is not quite a stright line but is noticeably "curved" in shape.


```python
%%R

X=seq(0,2,by=0.01)
set.seed(1)
X2 = 0.1*rnorm(length(X))
Y = 10*(X+X2)^2 - 8*(X+X2) + 1
plot(X, Y, xlab='X-axis',ylab='Y-axis',main=' \"Curved\" Shape of Data.')
res = lm(Y~X)
abline(res)
```

The straight line model doesn't "look like" a right fit.
But it does "look like" a quadratic equation plot.

We might try a "quadratic model" i.e. try to see if a we can fit  

$$Y = Ax^2 + Bx + C$$ 

to the data.

Here you might notice that we need to estimate *3* parameters.  This makes sense since our data has more variation or a more complex shape than a simple straight line.  So it's going to take more effort to find a curve that fits it.

In general as our model gets more sophisticated we need more coefficients to fit the model closely to the data, and so we need to do more work to find the right model.

Parameter Estimation by other methods 
-------------------------------------
The method of parameter estimation we are looking at above is called "Regression" which means we use the data to create formulas that give us the coefficients for the model.

This is not the only way to estimate the coefficients or parameters.

Other than Regressions there are stochastic techniques such as genetic programming and genetic algorithms which use evolutionary techniques.  In these approaches

but are more expensive in terms of computational resources and time needed. You might want to use these when the data sets are huge and conventional tools like R and MATLAB are not available at that scale.

#### Why would we want to create a model in this way?

We have some data and we want to use data to do provide some value. One way to do that it to predict what the variable might do outside the range where we can observe it - i.e. we want to predict from the data or extrapolate the data to areas where we may not be able to make observations.  We may use it to predict potential future values of a variable.

So Regression, in general, is a way to create a model from a set of observations, and then use it to make predictions, or to extrapolate from the given observations to other input values not observed.

Linear Regression is the special subset of Regression where the formula is a linear expression in the dependent variables, i.e. where our hypotheses is that Y is of the form Ax1 + Bx2 + Cx3 + D where A, B, C, D are constants. In general there may be any number of terms xn followed by a constant term.

In a matrix form this model can be expressed as 

$$Y = AX + B$$
