```python
Linear Regression - Data Exploration


[[ mention of Lending Club data set and some description of what Lending Club is etc ]]

A data exploration is done on the Lending Club sample in the following order

a) Browse the data - look at the variables and understand the data types
b) Data cleanup
c) Initial Visualization
d) Data segmentation
e) Summary Statistics


a) Browse Data
		 	 	 		
The data has the following variables (with explanation of meaning)
			
Amount.Requested - numeric
­‐ The amount (in dollars) requested in the loan application		

Amount.Funded.By.Investors - numeric
– The amount (in dollars) loaned to the individual

Interest.rate – character (a string of numbers with a % sign)
– The lending interest rate charged to the borrower.

Loan.length -- character (a string of numbers with “ months” appended)
– The length of time (in months) of the loan			

Loan.Purpose – categorical variable
– The purpose of the loan as stated by the applicant

Debt.to.Income.Ratio – character (a string of numbers with a % sign)
– The % of consumer’s gross income that goes toward paying debts 

State -- character (two letter variable)
– The abbreviation for the U.S. state of residence of the loan applicant Home.ownership -­‐ Indicates whether the applicant owns, rents, or has a mortgage 

Monthly.income -­ categorical
– The monthly income of the applicant (in dollars).

FICO.range  – categorical (expressed as a string label e.g. “650-655”)
– A range indicating the applicants FICO score. 	

Open.CREDIT.Lines -- numeric
– The number of open lines of credit at the time of application.

Revolving.CREDIT.Balance -- numeric
– The total amount outstanding all lines of credit 			

Inquiries.in.the.Last.6.Months -- numeric
– Number of  credit inquiries in the previous 6 months

Employment.Length -- character (a string of numbers with “ months” appended)
– Length of time employed at current job. 
		
We will need to transform character variables and most categorical variables to numeric.


b) Data Cleanup and typical issues

After cleaning the data and converting categorical variables to character and then numeric, the lower limit of the range can be used as the FICO.Score variable.

Cleaning involves
removing % signs from rates
removing the word ” months" from loan length.
managing outliers
managing NA

There is one especially high outlier with monthly income > 100K$+ This is likely to be a typo and is removed as a data item. There is also one data item with all N/A - this was also removed.

c) Initial Visualization
We plot Loan Interest rate as a function of FICO to see what this might tell us about the data distribution.
Why FICO - we know from domain knowledge that this is a major determinant of interest rate.
Figure 1. Notice that the two regions  (shaded blue) at each end show distinctly different distributions than the main region and also are distinct from the neighboring one at the boundary.  Hence in the modeling 3 different linear models were used, one for each region, to create a better fit.
```
